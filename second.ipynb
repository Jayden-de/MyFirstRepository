```python
# 컴퓨터와 대결하는 가위, 바위, 보 게임
# '그만'을 입력할때까지 계속
# '이겼습니다', '졌습니다', '비겼습니다' (전적: 승 패)

import random

sel = ['가위', '바위', '보']
input_sel = []
cnt = 0
win = 0
lose = 0
same = 0
while True:
    computer = random.randint(0, 2)
    computer = sel[computer]
    input_sel = input('가위바위보 선택: (종료하려면 "그만" 입력)')
    
    if input_sel == '그만':
        print('게임을 종료합니다.')
        print(f'전적 : {cnt}전, {win}승, {lose}패, {same}무승부')
        break
    
    if computer == input_sel:
        same += 1
        cnt += 1
        print('비겼다')
    elif (input_sel == '가위'): ## 가위
        if computer == '보':
            win += 1
            cnt += 1
            print('이겼다')
        else:
            lose += 1
            cnt += 1
            print('졌다')
    elif (input_sel == '바위'): ## 바위
        if computer == '가위':
            win += 1
            cnt += 1
            print('이겼다')
        else:
            lose += 1
            cnt += 1
            print('졌다')
    elif (input_sel == '보'): ## 보
        if computer == '바위':
            win += 1
            cnt += 1
            print('이겼다')
        else:
            lose += 1
            cnt += 1
            print('졌다')
```

```python
import random

sel = ['가위', '바위', '보']
input_sel = []
cnt = 0
win = 0
lose = 0
same = 0
log = []
while True:
    computer = random.randint(0, 2)
    computer = sel[computer]
    input_sel = input('가위바위보 선택: (종료하려면 "그만" 입력)')
    log.append(input_sel)

    if input_sel == '그만':
        print('게임을 종료합니다.')
        print(f'전적 : {cnt}전, {win}승, {lose}패, {same}무승부')
        print(log)
        break
    
    if input_sel not in sel:
         print('가위, 바위, 보 중에 선택 입력 하세요')
         continue

    if computer == input_sel:
        same += 1
        cnt += 1
        print('비겼다')
    elif (input_sel == '가위' and computer == '보')or(input_sel == '바위' and computer == '가위')or(input_sel == '보' and computer == '바위'):
            win += 1
            cnt += 1
            print('이겼다')
    else:
        lose += 1
        cnt += 1
        print('졌다')
```

```python
addr = {'박지민': '010-1111-1234', '전정국': '010-2222-1234', '민윤기': '010-3333-1234', '정호석': '010-4444-1234', '김태형': '010-5555-1234', '김석진': '010-6666-1234'}

while True:
    menu_sel = int(input('메뉴 선택(1: 추가, 2: 검색, 0: 종료)'))     

    if menu_sel == 0:
        print('프로그램을 종료합니다!')
        break

    elif menu_sel == 1:
        input_name = input('이름을 입력하세요..')
#        print(input_name)
        input_telnum = input('전화번호를 입력하세요..')
#        print(input_telnum)
        addr[input_name] = input_telnum

    elif menu_sel == 2:
        print(f'검색가능한 친구들: {list(addr.keys())}')
        input_name = input('이름을 입력하세요..')
#        print(input_name)
        if input_name in addr:
            print(f'{input_name}의 연락처 : {addr[input_name]}')
        else:
            print('등록되지 않은 친구입니다.')
            continue

    else:
        print(menu_sel)
        print('메뉴 선택 오류입니다. 다시 선택하세요.')
        continue
```

```python
morse = {
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f',
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l',
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r',
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x',
    '-.--':'y','--..':'z'
}
text = ''
letter = input('모스부호를 스페이스로 띄어쓰기하여 입력하세요.')
list_letter = letter.split()

for i in range(0, len(list_letter)):
    text += morse[list_letter[i]]

text
```
