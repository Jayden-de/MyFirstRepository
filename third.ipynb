```python
def num_max(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    else: max_val = a
    return max_val
```

```python
num_max(1, 12, 50)
```

```python
def num_min(a):
    min_val = a[0]
    for i in a:
        if i < min_val:
            min_val = i
    return min_val
```

```python
testarray = [52, 273, 32, 13, 275, 24, 57]
num_min(testarray)
```

```python
# 4로 나누어 떨어지는 해
# (그러나) 100으로 나누어 떨어지지 않는 해
# (그러나) 400으로는 나누어 떨어지는 해
def isLeapYear(year):
    return (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0)
```

```python
print(f'2022년은 윤년일까? {isLeapYear(2022)}')
print(f'2010년은 윤년일까? {isLeapYear(2010)}')
print(f'2000년은 윤년일까? {isLeapYear(2000)}')
print(f'1900년은 윤년일까? {isLeapYear(1900)}')
```

```python
def add_many(*args): # 가변인자
    result = 0
    for i in args:
        result += i
    return result
```

```python
print(add_many(1, 2, 3))
print(add_many(1, 2, 3, 4, 5, 6, 7, 8, 9))
```

```python
# 연산자까지 함께 받기
def add_mul(text, *args, opr=0):
    result = opr
    
    if text == '+':
        for i in args:
            result += i
    elif text == '*':
        result = 1
        for i in args:
            result *= i
    else: 
        print('연산자 입력 오류')
    return result
```

```python
print(add_mul('+', 4, 1, 2, 2, opr=1))
```

```python
def show_menu():
    print(f'=== 파이썬 카페 메뉴 ===')
    print(f'1. 아메리카노 (3000원)')
    print(f'2. 카페라떼 (4000원)')
    print(f'3. 종료하기')
    print(f'=======================')
    print()
```

```python
def calculate_change(money, price):
    if money > price:
        return print(f'주문이 완료되었습니다. 거스름돈은 {money-price}원 입니다.')
    elif money == price:
        return print(f'주문이 완료되었습니다. 거스름돈은 {money-price}원 입니다.')
    else:
        return print(f'금액이 {price-money}원 부족합니다.')
```

```python
menu = {'아메리카노': 3000, '카페라떼': 4000}
sp_menu = []

while True:
    show_menu()
    input_menu = input('menu 선택 띄우고 수량을 입력해 주세요')
    # 메뉴와 수량을 입력 후 분리
    sp_menu = input_menu.split()
    sel_menu = int(sp_menu[0])
    if sel_menu != 3:
        sel_count = int(sp_menu[1])
    # 메뉴 전처리
    if sel_menu == 3:
        print('키오스크를 종료합니다. 안녕히 가세요.')
        break
    elif sel_menu == 1:
        select = '아메리카노'
    elif sel_menu == 2:
        select = '카페라떼'
    # 메뉴 후처리
    price = menu[select] * sel_count
    print(f'{select}를 {sel_count}잔 선택하셨습니다. 가격은{price}원 입니다.')
    # 결재 처리
    input_money = int(input('지폐를 입력해주세요.'))
    calculate_change(input_money, price)
    print()
```

```python
def check_rule(prev_word, new_word, used_words):
    if prev_word == '':
        return True
    elif (new_word[0]==prev_word[-1]):
        for i in used_words:
            if i == new_word:
                return False
        return True
    else: return False
```

```python
print("=== 절대 안 봐주는 끝말잇기 게임 시작! ===")
print("(게임을 끝내려면 '그만'을 입력하세요)\n")

used_words = [] # 지금까지 쓴 단어를 저장할 빈 리스트
prev_word = ""  # 이전 단어를 저장할 변수

while True:
    if prev_word == "":
        new_word = input("첫 단어를 제시해 주세요: ")
    else:
        new_word = input(f"[{prev_word}] 다음 단어는? : ")

    # 중단 조건
    if new_word == '그만':
        print("게임을 포기하셨습니다!")
        break

    # 함수를 호출해서 규칙에 맞는지 검사 (결과는 True 또는 False)
    is_pass = check_rule(prev_word, new_word, used_words)

    if is_pass == True:
        print("통과!")
        used_words.append(new_word) # 통과한 단어를 '사용한 단어 목록'에 추가
        prev_word = new_word        # 다음 차례를 위해 '이전 단어'를 '지금 단어'로 업데이트
    else:
        # is_pass가 False라면 게임 오버
        print(f"게임 오버! 총 {len(used_words)}번 성공하셨습니다.")
        print(f"사용한 단어들: {used_words}")
        break
```
